```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid19_confirmed = pd.read_csv('COVID19_confirmed.csv')
covid19_deaths = pd.read_csv('COVID19_deaths.csv')
covid19_recovered = pd.read_csv('COVID19_recovered.csv')
```

```python
class Covid:
    def __init__(self, df, x , y):
        self.df = df
        self.x = x
        self.y = y
    def delete_extra_rows(self):
        self.df.drop(self.df.columns[0:4], axis=1, inplace=True)
        self.df.drop(self.df.columns[:344], axis=1, inplace=True)
    def delete_next_year(self):
        self.df.drop(self.df.columns[0:4], axis=1, inplace=True)
        self.df.drop(self.df.columns[345:], axis=1, inplace=True)
    def filter(self):
        self.df = self.df.loc[self.df['Country/Region'] == 'Romania']
    def difference(self):
        self.df = self.df.diff(axis=1)
    def plot(self, x, y):
        self.df = pd.DataFrame({x: self.df.columns[0:], 
                                y: self.df.iloc[0]})
        self.df = self.df[::10]
        plt.style.use('seaborn-white')
        plt.plot(self.df[self.x], self.df[self.y], marker='o')
        plt.grid(linestyle='--', linewidth=0.5)
        plt.xlabel(self.x)
        plt.ylabel(self.y)
        plt.rcParams["figure.figsize"] = (30,10)
```

## COVID19 confirmed cases in Romania from 2020 (detailed)

```python
covid19_confirmed_selected_2020 = Covid(covid19_confirmed, "Date", "Cases confirmed")
covid19_confirmed_selected_2020.filter()
covid19_confirmed_selected_2020.delete_next_year()
covid19_confirmed_selected_2020.difference()

covid19_deaths_selected_2020 = Covid(covid19_deaths, 'Date', 'Deaths')
covid19_deaths_selected_2020.filter()
covid19_deaths_selected_2020.delete_next_year()
covid19_deaths_selected_2020.difference()

covid19_recovered_selected_2020 = Covid(covid19_recovered, 'Date', 'Recovered')
covid19_recovered_selected_2020.filter()
covid19_recovered_selected_2020.delete_next_year()
covid19_recovered_selected_2020.difference()
```

```python
january_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,0:10].sum(axis=1)
february_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,10:39].sum(axis=1)
march_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,39:70].sum(axis=1)
april_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,70:100].sum(axis=1)
may_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,100:131].sum(axis=1)
june_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,131:161].sum(axis=1)
july_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,161:192].sum(axis=1)
august_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,192:223].sum(axis=1)
september_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,223:253].sum(axis=1)
october_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,253:284].sum(axis=1)
november_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,284:314].sum(axis=1)
december_2020_confirmed = covid19_confirmed_selected_2020.df.iloc[:,314:].sum(axis=1)

january_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,0:10].sum(axis=1)
february_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,10:39].sum(axis=1)
march_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,39:70].sum(axis=1)
april_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,70:100].sum(axis=1)
may_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,100:131].sum(axis=1)
june_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,131:161].sum(axis=1)
july_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,161:192].sum(axis=1)
august_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,192:223].sum(axis=1)
september_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,223:253].sum(axis=1)
october_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,253:284].sum(axis=1)
november_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,284:314].sum(axis=1)
december_2020_deaths = covid19_deaths_selected_2020.df.iloc[:,314:].sum(axis=1)

january_2020_recovered= covid19_recovered_selected_2020.df.iloc[:,0:10].sum(axis=1)
february_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,10:39].sum(axis=1)
march_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,39:70].sum(axis=1)
april_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,70:100].sum(axis=1)
may_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,100:131].sum(axis=1)
june_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,131:161].sum(axis=1)
july_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,161:192].sum(axis=1)
august_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,192:223].sum(axis=1)
september_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,223:253].sum(axis=1)
october_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,253:284].sum(axis=1)
november_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,284:314].sum(axis=1)
december_2020_recovered = covid19_recovered_selected_2020.df.iloc[:,314:].sum(axis=1)
```

## Cases of CORONA19 per month in Romania from 2020
